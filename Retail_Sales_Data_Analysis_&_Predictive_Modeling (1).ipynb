## Plot Monthly Revenue and Quantity Sold Trends

```python
import matplotlib.pyplot as plt
import seaborn as sns

plot_data = df.groupby([
    df['Date'].dt.year.rename('Year'),
    df['Date'].dt.month.rename('Month')
])[['Total Amount', 'Quantity']].sum().reset_index()

plot_data['Time'] = plot_data['Year'].astype(str) + '-' + plot_data['Month'].astype(str).str.zfill(2)
plot_data = plot_data.sort_values('Time')

sns.set_theme(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(12, 6))

#1- Plot the Total Amount on the primary Y-axis (Left side)
color = '#1f77b4'
ax1.set_xlabel('Timeline (Year-Month)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Total Revenue ($)', color=color, fontsize=12, fontweight='bold')
sns.lineplot(data=plot_data, x='Time', y='Total Amount', ax=ax1, color=color, marker='o', linewidth=2.5, label='Revenue')
ax1.tick_params(axis='y', labelcolor=color)
ax1.yaxis.set_major_formatter('${x:,.0f}') # Formats Y-axis values with currency markers

# 2. Create a secondary twin Y-axis tracking the Quantity metrics (Right side)
ax2 = ax1.twinx()
color = '#ff7f0e'
ax2.set_ylabel('Quantity Sold (Units)', color=color, fontsize=12, fontweight='bold')
sns.lineplot(data=plot_data, x='Time', y='Quantity', ax=ax2, color=color, marker='s', linewidth=2, linestyle='--', label='Quantity')
ax2.tick_params(axis='y', labelcolor=color)
ax2.yaxis.set_major_formatter('{x:,.0f}')

# Adjust layouts and finalize titles
plt.title('Monthly Sales Performance: Revenue vs. Quantity Trends', fontsize=15, fontweight='bold', pad=15)
fig.tight_layout()
plt.show()
```

#Modeling

```python
df["Price per Unit"].plot(kind="density")
plt.title("Density Plot")
plt.show()
```

```python
x=df[['Price per Unit', 'Quantity']]
y=df['Total Amount']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
```

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
reg_original=LinearRegression().fit(x_train,y_train)
print("Original Train R^2:",reg_original.score(x_train,y_train))

reg_scalled = LinearRegression()
reg_scalled.fit(x_train_scaled,y_train)
print("Scalled Train R^2:",reg_scalled.score(x_train_scaled,y_train))
```

```python
y_predict = reg_original.predict(x_test)
print(y_predict[0:5])

y_scaled_predict = reg_scalled.predict(x_test_scaled)
print(y_scaled_predict[0:5])
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
test_socre = r2_score(y_test,y_predict)
print(f"End-to-End R² Score For Unscaled Data: {test_socre}")
test_scaled_socre = r2_score(y_test,y_scaled_predict)
print(f"End-to-End R² Score For Scaled Data: {test_scaled_socre}")
test_mse = mean_squared_error(y_test, y_predict)
print(f"End-to-End Mean Squared Error For Unscaled Data: {test_mse}")
test_scaled_mse = mean_squared_error(y_test, y_scaled_predict)
print(f"End-to-End Mean Squared Error For Scaled Data: {test_scaled_mse}")
```

```python
print(reg_original.coef_)
print(reg_scalled.coef_)
```

```python
from mpl_toolkits.mplot3d import Axes3D
# Features
X1 = x_train['Price per Unit']
X2 = x_train['Quantity']
Y = y_train

# Create figure
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# Scatter points
ax.scatter(X1, X2, Y, color='red', alpha=0.6)

# Create mesh grid
x1_surf, x2_surf = np.meshgrid(
    np.linspace(X1.min(), X1.max(), 20),
    np.linspace(X2.min(), X2.max(), 20)
)

# Predict surface
y_surf = reg_original.predict(
    np.column_stack((x1_surf.ravel(), x2_surf.ravel()))
)

y_surf = y_surf.reshape(x1_surf.shape)

# Plot surface
ax.plot_surface(
    x1_surf,
    x2_surf,
    y_surf,
    alpha=0.4
)

# Labels
ax.set_xlabel('Price per Unit')
ax.set_ylabel('Quantity')
ax.set_zlabel('Total Amount')

ax.set_title('Multiple Linear Regression Plane')

plt.show()
```

## Business Insights and Recommendations

Based on the comprehensive analysis of the retail sales dataset, here are the key business insights and actionable recommendations:

### Key Business Insights:

1.  **Gender-Based Sales Performance:**
    *   **Total Sales & Quantity:** Female customers contribute slightly more to total sales ($232,840) and purchase a higher quantity of items (1298) compared to male customers (Sales: $223,160, Quantity: 1216).
    *   **Category Preference:** Females show stronger spending in 'Clothing' and 'Beauty' categories, while males lead in 'Electronics' sales.

2.  **Product Category Performance:**
    *   'Clothing' is the most frequently purchased category and also leads in total quantity sold.
    *   'Electronics' and 'Beauty' follow in terms of frequency and quantity.

3.  **Age Group Spending Patterns:**
    *   'Adults' (19-45 years) are the highest-spending demographic ($216,910), followed by 'Mature' customers (45-64 years) ($161,480), and then 'Teens' (17-19 years) ($87,610).
    *   Adults dominate sales across all product categories. Teens show significant spending in 'Clothing' and 'Beauty', while Mature customers also have notable spending in 'Clothing' and 'Electronics'.
    *   The correlation between age and total amount is very weak (-0.06), suggesting that age alone is not a strong predictor of individual transaction value.

4.  **Monthly Sales Trends:**
    *   Sales exhibit clear seasonality, with a notable peak in December (likely due to holiday shopping). There are also significant peaks in May and October.
    *   January 2024 data shows a sharp drop, primarily due to incomplete data for the month.

5.  **Customer Base:**
    *   The dataset contains 1000 unique customer IDs, indicating a diverse customer base for the given period.

### Recommendations:

1.  **Targeted Marketing Campaigns:**
    *   **Gender-Specific:** Develop marketing campaigns that specifically target female customers for 'Clothing' and 'Beauty' products. For male customers, focus promotional efforts on 'Electronics'.
    *   **Age-Specific:** Continue to focus marketing and product development efforts on the 'Adult' age group as they are the highest spenders. However, consider specialized promotions for 'Teens' in 'Clothing' and 'Beauty', and for 'Mature' customers in 'Clothing' and 'Electronics'.

2.  **Inventory Management & Promotions:**
    *   **Seasonal Stocking:** Prepare for increased demand in 'Clothing', 'Beauty', and 'Electronics' during peak months like May, October, and especially December.
    *   **Off-Peak Promotions:** Introduce special promotions or bundle offers during traditionally slower months (e.g., early Q1) to stimulate sales and maintain customer engagement.

3.  **Product Category Strategy:**
    *   Given 'Clothing' is the top-performing category, ensure a diverse and trending inventory. Invest in understanding the sub-trends within clothing for both male and female demographics.
    *   Explore opportunities to cross-sell 'Beauty' products with 'Clothing' for female customers, and 'Electronics' accessories for male customers.

4.  **Customer Loyalty Programs:**
    *   Since each customer ID appears only once, it suggests the dataset might capture individual transactions rather than repeat purchases. To foster loyalty, consider implementing a customer loyalty program to encourage repeat business and track customer lifetime value over time.
